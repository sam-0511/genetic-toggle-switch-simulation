```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import winsound
from ipywidgets import interact, FloatSlider

# -----------------------------
# TOGGLE SWITCH EQUATIONS
# -----------------------------
def toggle_switch(t, z, alpha1, alpha2, beta, gamma):
    u, v = z
    du_dt = alpha1 / (1 + v**beta) - u
    dv_dt = alpha2 / (1 + u**gamma) - v
    return [du_dt, dv_dt]

# -----------------------------
# MAIN SIMULATION FUNCTION
# -----------------------------
def run_sim(alpha1=2.5, alpha2=2.5, beta=2.0, gamma=2.0):
    t_eval = np.linspace(0, 50, 1000)

    # Two initial conditions
    sol1 = solve_ivp(toggle_switch, [0, 50], [8, 0.5],
                     args=(alpha1, alpha2, beta, gamma),
                     t_eval=t_eval)
    sol2 = solve_ivp(toggle_switch, [0, 50], [0.5, 8],
                     args=(alpha1, alpha2, beta, gamma),
                     t_eval=t_eval)

    # Final values
    final_u = sol1.y[0][-1]
    final_v = sol1.y[1][-1]
    print("Final U:", round(final_u, 3))
    print("Final V:", round(final_v, 3))

    # Sound
    if final_u > final_v:
        winsound.PlaySound(
            r"C:\Users\KIIT\Downloads\bombinsound-happy-kids-background-music-15-second-499550.wav",
            winsound.SND_FILENAME | winsound.SND_ASYNC)
    else:
        winsound.PlaySound(
            r"C:\Users\KIIT\Downloads\73750__timbre__remix-of-benboncan__sad_trombone_more_wah_brightde-clicked.wav",
            winsound.SND_FILENAME | winsound.SND_ASYNC)

    # Time series
    plt.figure(figsize=(10, 5))
    plt.plot(sol1.t, sol1.y[0], label='Protein U')
    plt.plot(sol1.t, sol1.y[1], label='Protein V')
    plt.xlabel('Time')
    plt.ylabel('Protein Concentration')
    plt.title('Genetic Toggle Switch')
    plt.legend()
    plt.grid(True)
    plt.savefig('toggle_timeseries.png', dpi=150)
    plt.show()

    # Phase portrait
    plt.figure(figsize=(5, 5))
    plt.plot(sol1.y[0], sol1.y[1], label='IC: U high', lw=2, color='green')
    plt.plot(sol2.y[0], sol2.y[1], label='IC: V high', lw=2, color='orange')
    plt.scatter([sol1.y[0][-1], sol2.y[0][-1]],
                [sol1.y[1][-1], sol2.y[1][-1]],
                s=80, zorder=5, label='Stable states')
    plt.xlabel('Protein U')
    plt.ylabel('Protein V')
    plt.title('Phase Portrait - Bistability')
    plt.legend()
    plt.savefig('toggle_phase.png', dpi=150)
    plt.show()

# -----------------------------
# INTERACTIVE SLIDERS
# -----------------------------
interact(
    run_sim,
    alpha1=FloatSlider(min=0.1, max=10, step=0.1, value=2.5),
    alpha2=FloatSlider(min=0.1, max=10, step=0.1, value=2.5),
    beta=FloatSlider(min=1, max=5, step=0.1, value=2),
    gamma=FloatSlider(min=1, max=5, step=0.1, value=2)
)
```
